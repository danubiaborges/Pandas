```python
import pandas as pd
```

```python
# dataframe = pd.DataFrame()
venda = {'data': ['15/02/2021', '16/02/2021'],
        'valor': [500, 300],
        'produto': ['feijao', 'arroz'],
        'qtde': [50, 70],
        }                             # dicionário 'venda'
vendas_df = pd.DataFrame(venda)       # criando dataframe (tabela)
```

```python
print(vendas_df)
```

```python
vendas_df = pd.read_excel("Vendas.xlsx")     # importando arquivo excel
display(vendas_df)
```

```python
# outros métodos de visualização:
display(vendas_df.head())    # só mostra as 5 primeiras linhas, se quiser mais linhas especificar dentro do parenteses
```

```python
display(vendas_df.shape)        # exibe a quantidade de linhas e colunas
```

```python
# métodos para editar o dataframe:
produtos = vendas_df['Produto']       # pegar 1 coluna e os pd.Series (uma coluna é uma serie do pandas)
display(produtos)
```

```python
produtos = vendas_df[['Produto', 'ID Loja']]     # quando tem 2 colunas não é mais uma serie e sim uma tablea
display(produtos)
```

```python
# método .loc para pegar 1 linha (utiliza os índices):
display(vendas_df.loc[1])           # uma única linha também é uma serie do pandas
```

```python
# pegar mais linhas com método .loc:
display(vendas_df.loc[1:5])
```

```python
# pegar linhas que correspondem a uma condição:
vendas_norteshopping_df = vendas_df.loc[vendas_df['ID Loja'] == 'Norte Shopping']
display(vendas_norteshopping_df)
```

```python
# pegar várias linhas e colunas usando o .loc (padrão .loc: [linhas, colunas])
vendas_norteshopping_df = vendas_df.loc[vendas_df['ID Loja'] == 'Norte Shopping', ['ID Loja', 'Produto', 'Quantidade']]
display(vendas_norteshopping_df)
```

```python
# pegar 1 valor específico com .loc:
# display(vendas_df.loc[linha, coluna])
display(vendas_df.loc[1, 'Produto'])
```

```python
# adicionar 1 coluna (duas formas):

# forma 1: a partir de uma coluna que já existe:
vendas_df['Comissão'] = vendas_df['Valor Final'] * 0.05
display(vendas_df)
```

```python
# criar uma coluna com valor padrão
vendas_df.loc[:, 'Imposto'] = 0      # : para o pandas significa todas as linhas
display(vendas_df)
```

```python
vendas_dez_df = pd.read_excel("Vendas - Dez.xlsx")     # imnportando vendas de dezembro
display(vendas_dez_df)
```

```python
# adicionar linhas
vendas_df = pd.concat([vendas_df, vendas_dez_df])
display(vendas_df)
```

```python
# excluir linhas e colunas
if 'Imposto' in vendas_df.columns:
    del vendas_df['Imposto']
```

```python
display(vendas_df)
```

```python
vendas_df = vendas_df.drop(0, axis=0)
display(vendas_df)
```

```python
# tratando valores vazios:

# deletar linhas e colunas completamente vazias
vendas_df = vendas_df.dropna(how='all', axis=1)
```

```python
# deletar linhas que possuem pelo menos 1 valor vazio
vendas_df = vendas_df.dropna()
```

```python
# preencher valores vazios (duas formas):
# preencher com a média da coluna:

vendas_df['Comissão'] = vendas_df['Comissão'].fillna(vendas_df['Comissão'].mean())
```

```python
display(vendas_df)
```

```python
# preencher com o último valor:
vendas_df = vendas_df.ffill()    # preenche com o valor acima
```

```python
# calcular indicadores (dois métodos):

# value counts
transacoes_loja = vendas_df['ID Loja'].value_counts()
display(transacoes_loja)
```

```python
# group by
faturamento_produto = vendas_df.groupby('Produto').sum(numeric_only=True)
display(faturamento_produto)
```

```python
faturamento_produto = vendas_df[['Produto', 'Valor Final']].groupby('Produto').sum(numeric_only=True)
display(faturamento_produto)
```

```python
faturamento_produto = vendas_df[['Produto', 'Valor Final']].groupby('Produto').mean()
display(faturamento_produto)
```

```python
# mesclar 2 dataframes (procurar informações de um dataframe em outro)
gerentes_df = pd.read_excel('Gerentes.xlsx')
display(gerentes_df)
```

```python
vendas_df = vendas_df.merge(gerentes_df)
display(vendas_df)
```
